```python
import time
import open3d
```

```python
model = open3d.io.read_point_cloud("objects/front/object_0.ply")
model2 = open3d.io.read_point_cloud("objects/front/object_1.ply")
```

```python
vis = open3d.visualization.Visualizer()
vis.create_window()
        
vis.add_geometry(model)
vis.add_geometry(model2)


while True:
    vis.poll_events()
    vis.update_renderer()
    time.sleep(0.05)

# vis.capture_screen_image(os.path.join(res_dir, f'vis_frame_{pred_id}.png'))
```
